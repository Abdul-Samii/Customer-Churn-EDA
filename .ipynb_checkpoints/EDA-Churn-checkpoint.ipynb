```python
missing = pd.DataFrame((df.isnull().sum())*100/df.shape[0]).reset_index()
plt.figure(figsize=(16, 8))
ax = sns.pointplot(x='index', y=0, data=missing)
plt.xticks(rotation = 90, fontsize = 7)
plt.title("Percentage of missing values")
plt.ylabel("PERCENTAGE")
plt.show()
```

#### Missing Data - Initial Intuition
- Here, we don't have any missing data.</br>
General Thumb Rules:
- For features with less missing values, we can use regression to predict the missing values or fill with the mean of the values present, depending upon the feature.</br>
- If you have more then 30%-40% missing values in a column you can delete it But you have to check for a dependency as well let suppose some other column(s) have some sort of dependency on the column you are going to drop. <b>For Example: </b>we have a column Car_Type which have some fields fill as SUV, HONDA but others are all null and there is another column is_Car which is a boolean column which shows if the person have car or not, so the people who do not have car have null Car_Type, so we cannot drop it, instead we replace the null with "no car".

### Data Cleaning

```python
new_df = df.copy()
```

<b>TotalCharges should be numeric amount. Let's convert this column to numeric.</b>

```python
new_df.TotalCharges = pd.to_numeric(new_df.TotalCharges, errors='coerce')
```

```python
new_df.info()
```

```python
new_df.isnull().sum()
```

```python
new_df.loc[new_df['TotalCharges'].isnull() == True]
```

<b>Missing Value Treatment</b></br>
Since the percentage of the missing values is less the 1 percent so we are going to delete the missing values.

```python
new_df.dropna(how = 'any', inplace=True)
```

#### For continous data like age, tenure we should perform feature binning 

```python
#Get the max tenure
print(new_df['tenure'].max()) #72 is max
```

```python
labels = ["{0} - {1}".format(i, i+11) for i in range(1, 72, 12)]
print(labels)
```

```python
new_df['tenure_group'] = pd.cut(new_df.tenure, range(1, 80, 12), right=False, labels=labels)
```

```python
new_df["tenure_group"].value_counts()
```

#### Removing some columns that are not impacting the target

```python
new_df.head()
```

```python
new_df.drop(columns=['customerID', 'tenure'], axis=1, inplace=True)
```

```python
new_df.head()
```

```python
#Data Cleaning Ended
```
